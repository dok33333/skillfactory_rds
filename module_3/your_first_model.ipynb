# Загрузка Pandas и очистка данных

```python
import pandas as pd
```

```python
df = pd.read_csv('main_task.csv')
df
```

```python
# Ваш код по очистке данных и генерации новых признаков
# При необходимости добавьте ячейки
```

```python
df['Number_of_Reviews_isNAN'] = pd.isna(df['Number of Reviews']).astype('uint8')
df['Number of Reviews'].fillna(0, inplace=True)
```

```python
df = pd.get_dummies(df, columns=[ 'City',], dummy_na=True)
```

```python
df.sample(5)
```

```python
df['Price Range'].value_counts()
```

```python
df['Price Range'].fillna("$$ - $$$", inplace=True)
df['Price Range'].value_counts()
```

```python
def replace_price_range(cell):
    if cell == '$':
        return 1
    elif cell == '$$ - $$$':
        return 2
    elif cell == '$$$$':
        return 3

df['Price Range'] = df['Price Range'].apply(replace_price_range)
df.sample(5)
```

```python
df['Cuisine Style'].fillna("['Other']", inplace=True)
df.sample(5)
```

```python
df['Cuisine Style'] = df['Cuisine Style'].apply(lambda x:x.replace('[', '').replace(']', '').replace("'", '').split(', '))
```

```python
df['Number_of_Cuisines'] = df['Cuisine Style'].apply(lambda x:len(x))
df.sample(5)
```

```python
cuisines_set = set()
def get_cuisines(cell):
    for item in cell:
        cuisines_set.add(item)
        
df['Cuisine Style'].apply(get_cuisines)
cuisines_set
```

```python
def find_cuisines(cell):
    if item in cell:
        return 1
    return 0

for item in cuisines_set:
    df[item] = df['Cuisine Style'].apply(find_cuisines)
df.sample(5)
```

```python
object_columns = [s for s in df.columns if df[s].dtypes == 'object']
object_columns = object_columns[1:]
```

```python
df.drop(object_columns, axis = 1, inplace=True)
```

```python
df.head(5)
```

```python
mean_ranking = round(df['Ranking'].mean(), 1)
df['Ranking'] = df['Ranking'].fillna(mean_ranking)
```

```python
mean_rating = round(df['Rating'].mean(), 1)
df['Rating'] = df['Rating'].fillna(mean_rating)
```

```python
mean_number_reviews = round(df['Number of Reviews'].mean(), 1)
df['Number of Reviews'] = df['Number of Reviews'].fillna(mean_number_reviews)
```

# Разбиваем датафрейм на части, необходимые для обучения и тестирования модели

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
X = df.drop(['Restaurant_id', 'Rating'], axis = 1)
y = df['Rating']
```

```python
# Загружаем специальный инструмент для разбивки:
from sklearn.model_selection import train_test_split
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Создаём, обучаем и тестируем модель

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
